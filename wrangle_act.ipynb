3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
#extracting each tweet status via their tweet_id in the twitter_archive data set
tweet_list=[]
tweet_list_err=[]
for ids in twitter_archive.tweet_id:
    try:
        tweet_list.append(api.get_status(ids, tweet_mode= 'extended'))
        print(ids,' : success')
    except tweepy.TweepError as e:
        tweet_list_err.append(ids)
        print (ids,' : failed')
```

```python
#list of tweet ids of tweets that failed to downloaded 
tweet_list_err
```

```python
#number of tweets that downloaded successfully
len(tweet_list)
```

```python
#creating a list 'json_data'
#storing the downloaded tweets in json_data
#reference: https://stackoverflow.com/questions/47612822/how-to-create-pandas-dataframe-from-twitter-search-api
json_data= []
for ids in tweet_list:
    json_data.append(ids._json)
```

```python
#storing the list in 'tweet_json.txt' file
#reference: https://stackoverflow.com/questions/47612822/how-to-create-pandas-dataframe-from-twitter-search-api
with open('tweet_json.txt', 'w') as file:
        file.write(json.dumps(json_data, indent=4))
```

```python
#creating a list 'retweet_fav_count'
#extracting the retweet count, favourite count, and ids from the 'tweet_json.txt' file
#appending the extracted data to the list "retweet_fav_count", and reading the data into a pandas dataframe 'json_data'
#reference: https://stackoverflow.com/questions/47612822/how-to-create-pandas-dataframe-from-twitter-search-api
retweet_fav_count = []
with open('tweet_json.txt', encoding='utf-8') as json_file:  
    json_tweet_data = json.load(json_file)
    for individual_tweets in json_tweet_data:
        tweet_id = individual_tweets['id']
        favorite_count = individual_tweets['favorite_count']
        retweet_count = individual_tweets['retweet_count']
        retweet_fav_count.append({'tweet_id': str(tweet_id),'favorite_count': int(favorite_count),'retweet_count': int(retweet_count)})
        json_data = pd.DataFrame(retweet_fav_count, columns = ['tweet_id', 'favorite_count','retweet_count'])
```

```python
#storing dataframe into csv file
json_data.to_csv('json_data.csv',index=False)
```
